# Data Visualization

I've picked the `Loan Data from Prosper` dataset to perform my data exploration and visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

## Dataset Overview

Below will be a high-level overview of the dataset. We will refer back to this information

```python
df = pd.read_csv('loanData.csv')
df.head()
```

For the purpose of this project, I will pick out the variables (columns) that I am interested in investigating from the dataset.
First I came up with an initial investigation question -
**What affects the borrowerâ€™s APR or interest rate?**

After reading through the dataset's data dictionary that explains each variables, and in order to answer the question and any further follow-up questions, I have pick the following variables from the dataset.

`IsBorrowerHomeowner`: A Borrower will be classified as a homowner if they have a mortgage on their credit profile or provide documentation confirming they are a homeowner.

`EmploymentStatus`: The employment status of the borrower at the time they posted the listing.

`IncomeRange`: The income range of the borrower at the time the listing was created.

`BorrowerRate`: The Borrower's interest rate for this loan. 

`LoanCurrentDaysDelinquent`: The number of days delinquent. 

`OnTimeProsperPayments`: Number of on time payments the borrower had made on Prosper loans at the time they created this listing. This value will be null if the borrower has no prior loans.

`LoanOriginalAmount`: The origination amount of the loan.


```python
loan_df = df[['EmploymentStatus','IsBorrowerHomeowner','IncomeRange','BorrowerRate','LoanCurrentDaysDelinquent','OnTimeProsperPayments','LoanOriginalAmount']]
print(loan_df.shape)
print(loan_df.dtypes)
```

To simplify the data exploration process, I will now convert the categorical variables into the categorical data type.

```python
# Convert EmploymentStatus and IncomeRange to ordered categorical data type
var_dict = {'EmploymentStatus': ['Employed','Full-time','Self-employed','Not available','Other','Part-time','Not employed','Retired'],
            'IncomeRange': ['$25,000-49,999','$50,000-74,999','$100,000+','$75,000-99,999','Not displayed','$1-24,999','Not employed','$0']}

for var in var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = var_dict[var])
    loan_df[var] = loan_df[var].astype(ordered_var)
```

## Univariate Exploration

I will start by looking at the main variable of interest: `BorrowerRate`. 

```python
# univariate plot of borrower rate
b = np.arange(0, loan_df.BorrowerRate.max() - 0.1, 0.01)
plt.hist(loan_df.BorrowerRate, bins = b);
plt.xlabel('Borrower Rate');
plt.title('Borrower Rate Frequency Distribution');
plt.ylabel('Frequency');
```

The distribution is a little bit skewed to the right and it is multimodal.

Now, I move to explore the 2 quality factors: `EmploymentStatus` and `IncomeRange`. 

```python
# univariate plots of EmploymentStatus and IncomeRange
plt.figure(figsize = [15, 5])
colorChoice = sb.color_palette()[0]

plt.subplot(1,2,1)
sb.countplot(data = loan_df, x = 'EmploymentStatus', color = colorChoice);
plt.ylim(0,70000);
plt.title('Employment Status frequency Distribution');
plt.ylabel('Frequency');
plt.xlabel('Employment Status');
plt.xticks(rotation=45);

plt.subplot(1,2,2)
sb.countplot(data = loan_df, x = 'IncomeRange', color = colorChoice);
plt.ylim(0,70000);
plt.title('Income Range frequency Distribution');
plt.ylabel('Frequency');
plt.xlabel('Income Range');
plt.xticks(rotation=45);
```

The 2 frequency distribution plots indicated that very rarely do unemployed or retired individual successful get a loan. 

## Bivariate Exploration

Now, I will investigate relationships between a pair of variables, especially how each of the variables relate to Borrower's rate.

Variables - The number of days delinquent, OnTimeProsperPayments, LoanOriginalAmount.

First I will create a plot of the borrower rate against The number of days delinquent.

```python
# bivariate plot of loan current days delinquent vs. borrower's rate
def log_trans(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

loan_df['log_LoanCurrentDaysDelinquent'] = loan_df.LoanCurrentDaysDelinquent.apply(log_trans)
    
sb.regplot(loan_df.LoanCurrentDaysDelinquent.apply(log_trans), loan_df.BorrowerRate, x_jitter = 0.3, scatter_kws = {'alpha' : 1/8});
plt.title('Log of Number of delinquent loan day vs. borrower\'s rate');
plt.xlabel('log of LoanCurrentDaysDelinquent');
plt.ylabel('Borrower\'s rate');
plt.ylim(0,0.5);
```

It is a surprise to see how many have loan delinquent for more than 1 day. However, it does not appear than it has significant enough impact on the borrower's rate.

```python
# bivariate plots of borrower's rate vs the number of days delinquent, OnTimeProsperPayments, LoanOriginalAmount
plt.figure(figsize = [15 , 4])
colorChoice = sb.color_palette()[0]

plt.subplot(1,2,1)
sb.violinplot(data = loan_df, x = 'EmploymentStatus', y = 'BorrowerRate', color=colorChoice);
plt.title('Income Range vs Borrower\'s rate');
plt.ylabel('Borrower\'s rate');
plt.xlabel('Employment Status');
plt.xticks(rotation=45);

plt.subplot(1,2,2)
sb.violinplot(data = loan_df, x = 'IncomeRange', y = 'BorrowerRate', color=colorChoice);
plt.title('Income Range vs Borrower\'s rate');
plt.ylabel('Borrower\'s rate');
plt.xlabel('Income Range');
plt.xticks(rotation=45);
plt.ylim(0,0.4)
```

It is interesting to see that the median borrower's rate is higher when employment status is not-employed, and higher frequency with borrowing higher rate for the not-employed category. 
Looking at the income range plot, i see a trend that median rate is lower when income is at the higher end.
